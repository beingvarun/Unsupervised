### import libraries 

```python
import pandas as pd
import dbscan_lab_helper as helper
from sklearn.cluster import DBSCAN
```

```python
dataset = pd.read_csv('blobs.csv')
```

```python
dataset_1 = dataset[:80].values
```

```python
helper.plot_dataset(dataset_1)
```

```python
dbscan = DBSCAN()
```

```python
labeled_data1 = dbscan.fit_predict(dataset_1)
```

```python
helper.plot_clustered_dataset(dataset_1, labeled_data1)
```

```python
helper.plot_clustered_dataset(dataset_1, labeled_data1, neighborhood=True)
```

```python
epsilon = 1.8
dbscan = DBSCAN(eps=epsilon)
clustered_label_data = dbscan.fit_predict(dataset_1)
helper.plot_clustered_dataset(dataset_1, clustered_label_data, neighborhood=True )
```
